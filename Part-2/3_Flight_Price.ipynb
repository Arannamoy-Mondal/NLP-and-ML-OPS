```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder 
```

```python
df=pd.read_excel("Dataset/002 flight-price.xlsx")
```

```python
df.head(5)
```

```python
encoder=OneHotEncoder()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Destination','Airline','Source']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df.drop_duplicates(inplace=True)
df[df.duplicated()]
```

```python
df['Journey_Date']=df['Date_of_Journey'].str.split("/").str[0].astype(int)
df['Journey_Month']=df['Date_of_Journey'].str.split("/").str[1].astype(int)
df['Journey_Year']=df['Date_of_Journey'].str.split("/").str[2].astype(int)
df.head(5)
```

```python
df['Dept_hour']=df['Dep_Time'].str.split(":").str[0]
df['Dept_minute']=df['Dep_Time'].str.split(":").str[1]
df.head(5)
```
