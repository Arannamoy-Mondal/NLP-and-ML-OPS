```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
hours=np.array(np.arange(2,12,2)).reshape(-1,1)
```

```python
marks=np.array([50,60,70,80,88])
```

```python
hours,marks
```

```python
model=LinearRegression()
```

```python
model.fit(hours,marks)
```

```python
display(model.coef_[0])
display(model.intercept_)
```

```python
# equation, y=mx+c
# here marks=y, m=coef, x=hour, c=intercept
```

```python
plt.scatter(hours,marks,color='blue',label='Actual Data')
predict_marks=model.predict(hours)
plt.plot(hours,predict_marks,color='red',label='Best Fit Line')
plt.grid(True)
plt.legend()
plt.ylabel("Marks")
plt.xlabel("hours")
```

```python
new_hours=np.array([7,8,9,10,11,12]).reshape(-1,1)
new_marks=model.predict(new_hours)
plt.scatter(new_hours,new_marks,color='blue',label='predict marks')
display(new_marks)
plt.legend()
plt.grid(True)
```
