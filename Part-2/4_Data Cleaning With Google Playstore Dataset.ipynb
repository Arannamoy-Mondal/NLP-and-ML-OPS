```python
df=df.drop(df.index[10472])
```

```python
df['Reviews']=df['Reviews'].astype(int)
```

```python
df['Size'].unique()
```

```python
# df['Size']=df['Size'].str.replace(".","")
df['Size']=df['Size'].str.replace("M","kk")
df['Size']=df['Size'].str.replace("k","")
df['Size']=df['Size'].replace("Varies with device",np.nan)
df['Size']=df['Size'].astype(float)
```

```python
df['Size'].head(5)
```

```python
df.info()
```

```python
df.head(1)
```

```python
character_to_remove=['+','$',',']
cols_to_clean=['Installs','Price']

for item in character_to_remove:
    for col in cols_to_clean:
        df[col]=df[col].str.replace(item,'')
```

```python
df['Installs'].unique()
```

```python
df['Price']=df['Price'].astype(float)
df['Installs']=df['Installs'].astype(int)
```

```python
display(df['Price'].unique())
display(df['Installs'].unique())
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
```

```python
df['Last Updated'].unique()
```

```python
df['Last Updated Day']=df['Last Updated'].dt.day
df['Last Updated Month']=df['Last Updated'].dt.month
df['Last Updated Year']=df['Last Updated'].dt.year
```

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv')
```

```python
numeric_cols=[feature for feature in df.columns if df[feature].dtype!='O']
categorical_cols=[feature for feature in df.columns if df[feature].dtype=='O']

display('numeric_cols: ',len(numeric_cols),numeric_cols)
display('categorical_cols: ',len(categorical_cols),categorical_cols)
```

```python
# Proportion of count in categorical column
for col in categorical_cols:
    display(col,'---->',df[col].value_counts(normalize=True)*100)
    display('----------------------') 
```

```python
# Proportion of count data on numerical columns

plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
for i  in range (0,len(numeric_cols)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df[numeric_cols[i]],fill=True,color='r')
    plt.xlabel(numeric_cols[i])
    plt.tight_layout()
```

### Observation
- Rating and Year is left skewed while Reviews, Size,Installs,Price are right skewed.  

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i  in range (0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
# Which is the most popular type category?
plt.figure(figsize=(15,16))
df['Category'].value_counts().plot.pie(y=df['Category'],figsize=(15,6),autopct="%1.1f")
plt.tight_layout()
```

```python
# Top 10 app category
category=pd.DataFrame(df['Category'].value_counts())
display(category[:10])
```

```python
plt.figure(figsize=(15,16))
sns.barplot(x=category.index[:10],y='count',data=category[:10],palette='hls')
plt.title("Top 10 app categories")
plt.xticks(rotation=90)
plt.show()
```

```python
# Q1:  Which Category  has largest number of installation?
apps=df.sort_values(by='Installs',ascending=False).head(1)
display(apps['Category'])
```

```python
# Q2: What are the most top 5 most installed apps in each popular categories ?
df_sorted=df.sort_values(by='Installs',ascending=False)
top_five_apps=df_sorted.groupby('Category').head(5)
display(top_five_apps[['App','Category','Installs']])
```
