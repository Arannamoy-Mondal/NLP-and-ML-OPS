````python
import pydantic 
from typing import List,Optional


class Student(pydantic.BaseModel):
    name:str=pydantic.Field(max_length=40,min_length=5)
    age:int
    city:str
    cpus:List[str]
    contact_no:Optional[str]=None
````

````python
print(Student.model_json_schema())
````
