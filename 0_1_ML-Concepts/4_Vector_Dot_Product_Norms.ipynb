```python
import numpy as np
```

```python
a=np.array([2,1,3])
b=np.array([1,3,3])
print(a)
print(b)
```

```python
add_ab=a+b
sub_ab=a-b
print(add_ab)
print(sub_ab)
```

### dot product

```python
dot_product=np.dot(a,b)
dot_pro=a@b #shortcut
print(dot_pro)
print(dot_product)
```

### norms


```python
l1_norm=np.linalg.norm(a,ord=1) # l1 norm length or magnitude  
print(l1_norm)
# l2 norm, the usual length
l2_a=np.linalg.norm(a)
print(l2_a)
```

```python
# normalizing unit vector
unit_a=a/np.linalg.norm(a)
l2_unia_a=np.linalg.norm(unit_a)
print(unit_a)
print(l2_unia_a)
```
