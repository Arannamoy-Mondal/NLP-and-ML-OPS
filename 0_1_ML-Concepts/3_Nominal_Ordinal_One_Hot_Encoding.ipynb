- Nominal: Categories with no order.
- Ordinal: Categories with order.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    "id":[1,2,3,4],
    "color":["red","blue","green","red"],
    "size":["Small","Medium","Large","Medium"],
    "price":[10,12,15,11]
})
```

### One hot encoding

```python
# Step-1: Apply one hot encoding to the color column
d_color=pd.get_dummies(df["color"],prefix="color",dtype=int)
```

```python
# Step-2: Combine the new column back with the original data
df_encoded=pd.concat([df,d_color],axis=1)
```

```python
# Step-3: Drop the old color column
df_encoded=df_encoded.drop("color",axis=1)
df_encoded
```

# Ordinal Encoding

```python
# declare the ordinal order
order={"Small":1,"Medium":2,"Large":3}
```

```python
# convert the whole feature
df["size_encoded"]=df['size'].map(order).astype(int)
```
