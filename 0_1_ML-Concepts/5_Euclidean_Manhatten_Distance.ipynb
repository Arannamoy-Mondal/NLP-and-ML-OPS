```python
import numpy as np
```

```python
x=np.array([
    [70,80],[60,90],[85,60],[78,76],[62,65]
],dtype=float)

q=np.array([75,70],dtype=float)
print("x shape: ",x.shape)
print("q shape:",q.shape)
```

### Euclidean distance

```python
eu=np.linalg.norm(x-q,axis=1)
print(eu.round(2).tolist())
```

### Manhatten distance

```python
man=np.linalg.norm(x-q,ord=1,axis=1)
print(man.tolist())
```
