```python
df=pd.read_csv("../Sample_Data/Daily_XAU_USD.csv")
df1=df.sort_index(ascending=0)
df1=df1.tail(10)
sns.relplot(kind="line",data=df1,y='high',x='datetime',aspect=10)
```

```python
pv=df.head(5).pivot(index="datetime",columns="high",values="high")
sns.heatmap(pv,annot=True,fmt=".2f",cmap="YlGnBu")
```

```python
df['datetime']=pd.to_datetime(df['datetime'])
```

```python
df['datetime']
```

```python
df['year']=df['datetime'].dt.year
```

```python
df
```

```python
yearly_max_high = df.groupby('year')['high'].max()
```

```python
yearly_max_high
```

```python
yearly_max_high=pd.DataFrame(data=yearly_max_high)
```

```python
yearly_max_high.rename({'price':'high'})
yearly_max_high
```

```python
sns.relplot(kind="line",data=yearly_max_high.head(3),y='year',x='high')
```

```python
sns.histplot(data=yearly_max_high.head(10),x='year',y='high')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=yearly_max_high,errorbar=None,x='year',y='high')
```
