### Central tendency: mean, median, mode


#### Mean:

- It is good when dataset has no outliers.

If mean greater than regular value then dataset has one or more outliers.

Example:

For [10,12,14,16,18,100]

mean=28.3, here 28.3 is greater than regular values except 100. So this dataset has outliers.

#### Median:
- It is good when dataset has outliers.
- Boxplot is best for median visualization.

Median is outlayer resistance.

Example:

For [10,12,14,16,18,100]

median=(14+16)/2=15, here 15 is almost equal to regular values except 100. So this dataset has outliers.

#### Mode:
- Most of them time used for categorical value like as color or browser.

|Situation|Best Measure|Reason|
|----------|-----------|----------|
|Symmetric data (height,core,thread)|Mean|All values contribute equally|
|Skewed data (income, price)|Mode|Resistance outliers|
|Categorical data ( color, browser, os )|Mode|non numeric data|


### Variance:
- average squared diviation of data points to mean.
- Two types of variance.
- 1. Sample variance. Bessel's correction used in sample variance. 
- 2. Population variance.

### Standard deviation:
- square root of variance

mean>median>mode --> right skewed ( positively skewed ) eg.. wealth, earning
  
mean < median < mode --> left skewed ( negatively skewed ) eg.. human life cycle

mean = median = mode --> symmetrical distribution ( machine loves it ) eg.. height

### Percentile:
- Q2-Q1 = Q3-Q2 ,the data is symmetric.
- Q2-Q1 < Q3-Q2 ,the data is right-skewed (the upper 25% is more spread out).
- Q2-Q1 > Q3-Q2 ,the data is left-skewed (the lower 25% is more spread out). 
- IQR=Q3-Q1
- Outlier range: Q1-1.5*IQR -----> Q3-1.5 *IQR

#### Long tailed data challenges ML models:
 Models may overfit  to frequent cases and ignore rare but important events - like fraud detection or rare disease prediction. Handling long tails often involves data normalization, resampling or outlier-aware models.
