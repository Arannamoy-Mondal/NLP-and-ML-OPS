```python
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader=TextLoader("speech.txt")
data=loader.load()
data[0].page_content
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
split=text_splitter.split_documents(data)
```

```python
split
```

```python
embedding=OllamaEmbeddings(model="phi4:latest")
vectordb=Chroma.from_documents(documents=split,embedding=embedding,persist_directory="chromadb")
vectordb
```

```python
query="calculate"
similarity_search=vectordb.similarity_search(query=query)
```
