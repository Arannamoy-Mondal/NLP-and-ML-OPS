### FAISS -> Facebook AI Similarity Serach 

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
```

```python
documents=TextLoader("speech.txt").load()
```

```python
docs=CharacterTextSplitter(chunk_size=200,chunk_overlap=0).split_documents(documents)
```

```python
embedding=OllamaEmbeddings(model="phi4:latest")
```

```python
db=FAISS.from_documents(docs,embedding)
```

```python
db
```

```python
### query
query1="Calculate this math."
docs=db.similarity_search(query=query1)
docs[0].page_content
```

```python
retriever=db.as_retriever()
```

```python
docs=retriever.invoke(query1)
docs[0].page_content
```

```python
docs_similarity_score=db.similarity_search_with_score(query=query1)
docs_similarity_score
```

```python
db.save_local("faiss_data")
```

```python
new_db=db.load_local("faiss_data",embedding,allow_dangerous_deserialization=True)
```

```python
docs=new_db.similarity_search(query=query1)
```
